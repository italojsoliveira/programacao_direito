## Para acessar e programar no Colab

Para acessar ao Google Colab e acessar a uma opção de executar o Jupyter online, acesse o link:

https://colab.research.google.com

Você provavelmente encontrará essa página:


Caso esteja deslogado, clique no botão "Sign in" na parte superior, à direita para logar com a sua conta da Google.



Em seguida, clique em "File", e depois em "New Python 3 Notebook".

Uma nova aba irá abrir com o novo arquivo Jupyter para edição. A tela se apresentará dessa forma:

## Para visualizar os arquivos .ipynb

Os arquivos .ipynb estarão salvos na pasta "Colab Notebooks" do Google drive. Para acessá-los, basta acessar:
    
https://drive.google.com/

Em seguida, clicar duas vezes em "My Drive" e depois, clicar em "Colab Notebooks". O arquivo salvo aparecerá nesse diretório como esperado. Esse arquivo .ipynb é o arquivo que deve ser submetido para avaliação.
